```python
def txsAvg(txs):
    gYear=txs.groupby(['year'])
    yearly_avgs= gYear.aggregate({"sales":np.mean})
    gMonth=txs.groupby(['year','month'])
    monthly_avgs= gMonth.aggregate({"sales":np.mean})
    gWeek=txs.groupby(['year','week_number'])
    weekly_avgs=gWeek.aggregate({"sales": np.mean})
    return yearly_avgs,monthly_avgs,weekly_avgs
```
